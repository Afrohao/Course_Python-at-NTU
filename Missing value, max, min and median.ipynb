```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jan 21 14:09:48 2019

@author: hao-ling.weng
"""

import pandas as pd 
df_hw1= pd.read_csv("sonar_hw1.csv")

df_hw1 = df_hw1.drop('object', axis=1)
n_cases = df_hw1.shape[0] # Number of cases
missing = n_cases - pd.DataFrame(df_hw1.count(), columns=['Missing'])

df_low= pd.DataFrame(df_hw1[df_hw1<0].count(), columns=['Low Outliers'])
df_hi= pd.DataFrame(df_hw1[df_hw1>1].count(), columns=['Hi Outliers'])

df_accept = df_hw1[(df_hw1>=0) & (df_hw1<=1)]

minimum = pd.DataFrame(df_accept.min(), columns=['Min'])
maximum = pd.DataFrame(df_accept.max(), columns=['Max'])
median = pd.DataFrame(df_accept.median(), columns=['Median'])


df_list = [df_low, df_hi, minimum, median, maximum]

df = missing

for df_ in df_list:
    df = df.join(df_)
print(df)
```
